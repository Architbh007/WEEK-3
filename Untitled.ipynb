```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the CSV file
df = pd.read_csv("natural_random_walk_sensor_data.csv")

# Step 2: Inspect the first few rows of the data to confirm it's loaded correctly
print("Data Sample:")
print(df.head())

# Step 3: Convert the 'timestamp' column to datetime format, ignoring any errors
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

# Step 4: Verify that there are no NaN values in the timestamp or acceleration columns
print("\nMissing values in each column:")
print(df.isnull().sum())

# Step 5: Drop any rows with missing values in the timestamp or acceleration columns
df_cleaned = df.dropna(subset=['timestamp', 'acceleration_x', 'acceleration_y', 'acceleration_z'])

# Step 6: Print the cleaned data to ensure it's ready for plotting
print("\nCleaned Data Sample:")
print(df_cleaned.head())

# Step 7: Plot the data for each axis

# Plot acceleration_x over time
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['timestamp'], df_cleaned['acceleration_x'], label='Acceleration X', color='blue')
plt.title('Acceleration in X over Time')
plt.xlabel('Timestamp')
plt.ylabel('Acceleration (g)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot acceleration_y over time
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['timestamp'], df_cleaned['acceleration_y'], label='Acceleration Y', color='orange')
plt.title('Acceleration in Y over Time')
plt.xlabel('Timestamp')
plt.ylabel('Acceleration (g)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot acceleration_z over time
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['timestamp'], df_cleaned['acceleration_z'], label='Acceleration Z', color='green')
plt.title('Acceleration in Z over Time')
plt.xlabel('Timestamp')
plt.ylabel('Acceleration (g)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Combined plot for acceleration_x, acceleration_y, acceleration_z
plt.figure(figsize=(10, 6))
plt.plot(df_cleaned['timestamp'], df_cleaned['acceleration_x'], label='Acceleration X', color='blue')
plt.plot(df_cleaned['timestamp'], df_cleaned['acceleration_y'], label='Acceleration Y', color='orange')
plt.plot(df_cleaned['timestamp'], df_cleaned['acceleration_z'], label='Acceleration Z', color='green')
plt.title('Combined Accelerometer Data Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Acceleration (g)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
